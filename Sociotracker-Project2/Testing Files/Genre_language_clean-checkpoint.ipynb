```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Extract Netflix Original Movies
movies_file = "../Resources/NetflixFilm Originals.csv"
movies_data_df = pd.read_csv(movies_file)
movies_data_df.head()
```

```python
movies_langauge_df=(movies_data_df.set_index(['Title','Genre','Premiere', 'Runtime', 'IMDB Score'])
   .apply(lambda x: x.str.split('/').explode())
   .reset_index())    
movies_langauge_df.head()
```

```python
movies_genre_df=(movies_langauge_df.set_index(['Title','Premiere', 'Runtime', 'IMDB Score','Language'])
   .apply(lambda x: x.str.split('/').explode())
   .reset_index())  
movies_genre_df.head()
```
