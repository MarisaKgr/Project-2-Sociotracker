```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Extract Stock Prices data
stock_file = "../Resources/Yahoo finance NFLX.csv"
stock_data_df = pd.read_csv(stock_file)
stock_data_df.head()
```

```python
#Make new df to add quarter data
stock_data_quarter_df=pd.DataFrame(stock_data_df)
stock_data_quarter_df
```

```python
stock_data_quarter_df["Quarter_ID"]= pd.PeriodIndex(stock_data_df.Date, freq='Q')
```

```python
stock_data_quarter_df['Date'] = pd.to_datetime(stock_data_df['Date'], format='%m/%d/%Y')
```
